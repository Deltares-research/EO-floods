```python
import geopandas as gpd
from shapely.geometry import Point
from functions import date_checker
```

This notebook is for checking if for a given date and region icesat2 data is available.
In order to check this, you need a region bounding box in the format of xmin, ymin, xmax, ymax coordinates in epsg 4326
and a date in the format of 'dd-mm-yyyy'.
The date_checker function checks for the given date and a day before and after. This increases the chance of an overlap of icesat2 data and a flooded area.

```python
# Example
dates = ["19-10-2019", "31-10-2019", "15-01-2021", "08-07-2021", "31-08-2021",  "02-08-2021", "10-11-2020", "21-08-2021"] # you can supply multiple dates
aoi = [109.03539316164964,13.259889529508994,109.29494516360276,13.405209730684632 ] # bounding box of area of interest
df = date_checker(aoi, dates=dates)
df
```

```python
# Create a geodataframe from the dataframe for visualization
geometry = [Point(xy) for xy in zip(df.lon, df.lat)]
df.drop(["lon", "lat"], axis=1)
gdf = gpd.GeoDataFrame(df, crs=4326, geometry=geometry)
gdf.explore()
```
